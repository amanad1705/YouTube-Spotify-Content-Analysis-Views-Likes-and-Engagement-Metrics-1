#  Comparison of Album Popularity on Spotify 

```python
avg_alb = df.groupby('Album_type')[["Views","Likes","Comments"]].mean().reset_index()

avg_alb
```

```python
avg_alb_melt = avg_alb.melt(
    id_vars="Album_type",       # columns to keep
    value_vars=["Views","Likes","Comments"],  # columns to melt
    var_name="Metric",          # new column name for melted variable
    value_name="Average"        # new column name for values
)

avg_alb_melt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data=avg_alb_melt, x="Album_type", y="Average", hue="Metric")
plt.title("Average Views, Likes, and Comments per Album Type")
plt.show()
```

#  Top YouTube Channels by Total Views

```python
c_view = df.groupby('Channel')["Views"].sum().sort_values(ascending=True)
top_5_channels=c_view.head(5)
```

```python
import matplotlib.pyplot as plt

# Top 5 channels by views
top_5_channels = c_view.sort_values(ascending=False).head(5)

plt.figure(figsize=(10,6))
plt.barh(top_5_channels.index, top_5_channels.values, color="#66b3ff")
plt.title("Top 5 YouTube Channels by Total Views")
plt.xlabel("Total Views")
plt.ylabel("Channel")
plt.gca().invert_yaxis()  # optional: highest at top
plt.tight_layout()
plt.show()
```
